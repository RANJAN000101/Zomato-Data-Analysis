```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe=pd.read_csv('/content/drive/MyDrive/Zomato data .csv')
dataframe.head()
```

```python
dataframe.tail()
```

**CONVERT THE DATA TYPE OF COLUMN-RATE**

```python
def handleRate(value):
  value=str(value).split('/')
  value=value[0]
  return float(value)
dataframe['rate']=dataframe['rate'].apply(handleRate)
dataframe.head()
```

```python
dataframe.info()
```

**1.COUNT THE TYPES OF RESTAURANT WHERE CUSTOMERS ORDER FROM**

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('Type of Restaurant',size=20)
```

```python
dataframe.head()
```

**2.NUMBER OF VOTES RECEIVED FROM CUSTOMERS**

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result,c='green',marker='o')
plt.xlabel('Type of Restaurant',size=20,c='red')
plt.ylabel('Votes',size=20,c='red')
```

**3.THE RATINGS RECEIVED FROM MAJORITY OF RESTAURANTS**

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins =5)
plt.xlabel('Ratings',size=20)
plt.title('Ratings Distribution',size=20)
plt.show()
```

**4.AVERAGE ORDER SPENDING BY COUPLES**

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
plt.show()
```

**5.WHICH MODE (ONLINE/OFFLINE) RECIEVES MAXIMUM RATING**

```python
dataframe.head()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=dataframe['online_order'],y='rate',data=dataframe)
plt.show()
```

**6.TYPE OF RESTAURANTS RECEIVED MORE OFFLINE ORDERS**

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap='YlGnBu',fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In(Type)")
plt.show()
```
